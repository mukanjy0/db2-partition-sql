```python
import matplotlib.pyplot as plt

def get_graph(x, y1, y2, nx, ny1, ny2, filename=None):
  fig, ax = plt.subplots()

  ax.plot(x, y1, marker='o', color='purple', linestyle='-')
  ax.plot(x, y2, marker='o', color='green', linestyle='-')
  ax.set_yscale('log')

  ax.legend([ny1, ny2])

  ax.set_xlabel(nx)
  ax.set_ylabel('Tiempo (ms)')
  ax.set_title(f'Query versus tiempo (ms)')

  plt.show()
  if filename:
    fig.savefig("vis/" + filename)
```

# P1

```python
queries = ['Q1', 'Q2', 'Q3']
employees = [605.311, 171.956, 87.348]
employees1 = [172.772, 33.744,  26.785]
get_graph(queries, employees, employees1, "Query", "employees", "employees1", filename="p1.png")
```
